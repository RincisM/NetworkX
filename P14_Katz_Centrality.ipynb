```python
import networkx as nx
import math
```

```python
#Creating a Bipartite Graph with 2 sets of nodes
B = nx.DiGraph()

user = ["A", "B", "C", "D"]
items = [10, 20, 30]

B.add_nodes_from(user, bipartite=0)
B.add_nodes_from(items, bipartite=1)
B.add_edges_from([("A", 10), ("B", 30), ("C", 10), ("A", 20), ("B", 20), ("C", 10), ("D", 30), ("D", 20)])
pos = nx.bipartite_layout(B, user)

nx.draw(B, pos, with_labels=True)
```

```python
#Transposing the Directed Graph
G = nx.DiGraph.reverse(B)

nx.draw(G, pos, with_labels=True)
```

```python
katz_centrality = nx.katz_centrality(G)
for node, centrality in katz_centrality.items():
    print(f"{node}: {centrality}")
```
