```python
import networkx as nx
```

```python
#Defining the Vertices and edges
vertices = range(1, 4)
edges = [(1, 2), (2, 3)]
```

```python
G = nx.Graph()
```

```python
#Drawing the Graph
G.add_nodes_from(vertices)
G.add_edges_from(edges)

nx.draw(G, with_labels=True, font_weight='bold')
```

```python
#Adding Nodes and Edges and using nx.draw to plot the graph

G.add_node(4)
G.add_node(5)
G.add_edge(2, 4)

nx.draw(G, with_labels=True, font_weight='bold')
```
