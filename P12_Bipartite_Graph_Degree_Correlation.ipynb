```python
import networkx as nx
```

```python
#Creating a Bipartite Graph with 2 sets of nodes
B = nx.Graph()

user = ["A", "B", "C", "D"]
items = [10, 20, 30]

B.add_nodes_from(user, bipartite=0)
B.add_nodes_from(items, bipartite=1)
B.add_edges_from([("A", 10), ("B", 30), ("C", 10), ("A", 20), ("B", 20), ("C", 10), ("D", 30), ("D", 20)])
pos = nx.bipartite_layout(B, user)

nx.draw(B, pos, with_labels=True)
```

```python
#Closeness Centrality
centrality_data = nx.closeness_centrality(B)

print(centrality_data)

max_key = max(centrality_data, key=centrality_data.get)
max_value = max(centrality_data.values())
print("Node with Maximum Closeness Centrality\n", max_key, ":", max_value)
```

```python
#Degree Correlation
print(nx.degree_assortativity_coefficient(B))
```
