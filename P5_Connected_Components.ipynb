```python
import networkx as nx
```

```python
#Defining the Vertices and edges
vertices = range(1, 4)
edges = [(1, 2), (2, 3)]
```

```python
G = nx.Graph()
```

```python
#Drawing the Graph
G.add_nodes_from(vertices)
G.add_edges_from(edges)

nx.draw(G, with_labels=True, font_weight='bold')
```

```python
#Adding Nodes and Edges and using nx.draw to plot the graph

G.add_node(4)
G.add_node(5)
G.add_edge(2, 4)

nx.draw(G, with_labels=True, font_weight='bold')
```

```python
#Finding the Degree of the Graph

degree_of_graph = nx.degree(G)
print(degree_of_graph)
```

```python
#Plotting the Degree Distribution

import matplotlib.pyplot as plt

nodes, degree = zip(*degree_of_graph)

plt.hist(degree)
plt.xlabel("Degree")
plt.ylabel("No. of Nodes")
plt.title("Degree Distribution")
plt.show()
```

```python
plt.bar(nodes, degree)
plt.xlabel("Nodes")
plt.ylabel("Degree")
plt.title("Degree Distribution")
plt.show()
```

```python
#Finding the Shortest Path

import pandas as pd
shortest_path = nx.shortest_path(G, method='dijkstra')
s_p_data = pd.DataFrame(shortest_path)
print(s_p_data)
```

```python
#Finding the Shortest Path between 1 and 4
shortest_path_1_4 = nx.shortest_path(G, source=1, target=4,method='dijkstra')

shortest_path_1_4
```

```python
#Finding the Connected Components
[c for c in sorted(nx.connected_components(G), key=len, reverse=True)]  #Results represent the nodes in each component - in our graph, there
                                                                        #are two components, one with 4 nodes and the other with 1 node sorted in descending
                                                                        #order based on the number of nodes in each component
```
